### **Group 1 → Identifiers & Basic Info**
##### (ID, Payment, City, gender, Age, Nature_Sign)

#### **a) ID, Payment, City, gender, Age, Nature_Sign (Columns):-**

##### **Objective:-**

##### To Evaluate The Relevance Of Columns `ID, Payment, City, Gender, Age, And Nature_Sign` For The Analysis And Decide Which To Retain Or Drop.

##### Remove Redundant Or Non-Informative Fields `(ID, City, Nature_Sign)` To Simplify The Dataset And Prevent Accidental Use In Modeling.

##### **1. Objective**
##### The Industry Column Contains Information About The Sector Or Domain In Which Customers Are Employed Or Associated. It Is A Categorical Variable But Contains Duplicate Entries, Inconsistent Cases, Spelling Variations, And Special Characters. ##### **Case Sensitivity:** Some Values In Lowercase, Uppercase, Or Mixed (E.g., Software & IT Services, software and it services).

##### **Special Characters & Spaces:** Extra Spaces (BFSI ) Or Symbols (Education & Training, education and training).

```python
import re

brand_mapping = {
    "iphone": "Apple IPhone",
    "ipad": "Apple IPad",
    "asus": "Asus",
    "google": "Google Pixel",
    "honor": "Honor",
    "huawei": "Huawei",
    "infinix": "Infinix",
    "itel": "Itel",
    "lava": "Lava",
    "lenovo": "Lenovo",
    "lg": "LG",
    "micromax": "Micromax",
    "motorola": "Motorola",
    "nokia": "Nokia",
    "nothing": "Nothing",
    "oneplus": "OnePlus",
    "oppo": "Oppo",           
    "poco": "POCO",
    "realme": "Realme",
    "redmi": "Redmi",
    "samsung": "Samsung",
    "tecno": "TECNO",
    "vivo": "Vivo",
    "xiaomi": "Xiaomi",
    "iqoo": "iQOO",
}

def clean_device_model(model):
    """Standardize smartphone brand names from the device_model column."""
    if pd.isna(model) or str(model).strip() == "":
        return "NaN"

    text = str(model).strip()
    # Take The Part Before '::', Lowercase It
    prefix = text.split("::")[0].lower()

    for key, value in brand_mapping.items():
        if prefix.startswith(key):
            return value

    return text

# Apply
data["device_model"] = data["device_model"].apply(clean_device_model)

print(sorted(data["device_model"].unique()))
```

#### **b) Age_Bucket (Columns):-**
##### **Objective:-**

##### Review Existing Age_Bucket Labels And Correct Unclear Or Incomplete Values (E.g., “45 Yrs & Above”).
##### Create Clear, Non Overlapping Buckets To Support Accurate Segmentation And Analysis.

##### **Observation:-**

##### Current Age_bucket Values Are Inconsistent, E.g.:

##### 24 Yrs And Below
##### 25 Yrs To 34 Yrs
##### 35 Yrs To 44 Yrs
##### 45 Yrs & Above (Ambiguous Upper Bound)

##### Missing Values May Also Exist In Some Records.

##### Need To Rename And Align Buckets With Clear Numerical Ranges.

##### **Proposed Standard Bins**
##### **(Corrected Label)**	 
##### 24 Yrs And Below	 
##### 25 Yrs To 34 Yrs	 
##### 35 Yrs To 44 Yrs	 
##### 45 Yrs To 60 Yrs	 
##### 60 Yrs & Above	 

```python
def clean_age_bucket(val):
    """
    Standardize age groups in Age_Bucket.
    """
    if pd.isna(val) or str(val).strip() == "":
        return np.nan   # Keep NaN For Now If Needed

    text = str(val).strip().lower()

    if "24" in text and "below" in text:
        return "24 yrs and below"
    if "25" in text and "34" in text:
        return "25 yrs to 34 yrs"
    if "35" in text and "44" in text:
        return "35 yrs to 44 yrs"
    if text.startswith("45"):
        return "45 yrs to 60 yrs"
    if text.startswith("60"):
        return "60 yrs & Above"

    return val  # Fallback If Unexpected Value

# Apply The Function
data["Age_Bucket"] = data["Age_Bucket"].apply(clean_age_bucket)

print("Unique Age Buckets After Cleaning:")
print(sorted(data["Age_Bucket"].unique()))
```

#### **c) Standardizing & Aggregating Investment (Columns):-**
##### [`RE_Residential, RE_Commercial, Equity_MF, Equity_ESOPs, Equity_Stocks, Debt_EPF, Debt_FD, Debt_GovtSchemes, Debt_HybridMF, Debt_MF, Debt_NPS, Debt_Other, Debt_PPF, Debt_Saving, Gold_Digital, Gold_Physical, Gold_SGB, AltInv_Crypto, AltInv_P2P_OR_LoansGiven, AltInv_Silver`]

##### **Objective:-**

##### Consolidate Granular Asset Fields (E.g., Equity_MF, Debt_FD, Gold_Physical, Etc.) Into Five Master Columns:

##### `-> Real_Estate`

##### `-> Equity`

##### `-> Debt`

##### `-> Gold`

##### `-> Alt_Investments`

##### Replace Blank / Null Values With 0 For Consistency.

##### Convert Values Into Millions With Three Decimals To Simplify Analysis And Reporting.

##### Remove Redundant Sub-Columns After Aggregation.

##### **Observation:-**

##### Many Asset Categories Were Split Across Multiple Columns, Creating Redundancy And Complicating Analysis.

##### Several Records Contained Missing Or Blank Values, Which Could Distort Totals If Not Handled.

##### **After Aggregation:-**

##### All Sub-Columns Successfully Merged Into Their Respective Category Columns.

##### Values Are Now Standardized In Million-Decimal Format (E.g., 45.000 = 45M).

##### Simplified Structure Makes It Easier To Compare Asset Classes And Validate Against Total_assets.

##### The Dataset Is Now More Compact And Ready For Modeling, Dashboards, Or Reporting.

```python
# a) Defined Groups:-

# Real Estate, Equity,Debt, Gold, Alternative Investments

real_estate_cols = ['RE_Residential', 'RE_Commercial']
equity_cols = ['Equity_MF', 'Equity_ESOPs', 'Equity_Stocks']
debt_cols = ['Debt_EPF','Debt_FD','Debt_GovtSchemes','Debt_HybridMF',
             'Debt_MF','Debt_NPS','Debt_Other','Debt_PPF','Debt_Saving']
gold_cols = ['Gold_Digital','Gold_Physical','Gold_SGB']
alt_cols = ['AltInv_Crypto','AltInv_P2P_OR_LoansGiven','AltInv_Silver']

# b) Convert To Numeric & Replace Blanks With 0:-

data[real_estate_cols + equity_cols + debt_cols + gold_cols + alt_cols] = (
    data[real_estate_cols + equity_cols + debt_cols + gold_cols + alt_cols]
      .apply(pd.to_numeric, errors='coerce')
      .fillna(0)
)
```

```python
# c) Create One Column Per Category (Sum Row Wise)

data["Real_Estate"]   = data[real_estate_cols].sum(axis=1)
data["Equity"]        = data[equity_cols].sum(axis=1)
data["Debt"]          = data[debt_cols].sum(axis=1)
data["Gold"]          = data[gold_cols].sum(axis=1)
data["Alt_Investments"] = data[alt_cols].sum(axis=1)

# d) Convert To Millions With 3 Decimals

cols_final = ["Real_Estate","Equity","Debt","Gold","Alt_Investments"]

data[cols_final] = (data[cols_final] / 1_000_000).round(3)
pd.set_option('display.float_format', '{:.3f}'.format)
print(data[cols_final].head())
```

```python
# e) Drop The Original Granular Columns:-

data.drop(columns=real_estate_cols + equity_cols + debt_cols + gold_cols + alt_cols, inplace=True)
```

### **Group 6 → Liabilities / Loans**
##### (Total_Liabilities, Housing_Loan, Auto_Loan, Personal_Loan, Consumer_Loan, Credit_Card, Education_Loan, Gold_Loan, Other_loan)

#### **a) Liabilities & Loan Data – Consolidation And Cleaning:-**

##### **Objective:-**

##### Review All Columns Related To Customer Liabilities And Loans.

##### Remove Redundant Column Total_liabilities (Its Value Equals The SUM Of Individual Loan Types).

##### Replace Any NaN / Blank Values With 0 For Accurate Numeric Analysis.

##### Merge Sub-Categories That Represent The Same Type Of Loan Into A Single Column (E.g., Any Variants Of “Personal Loan”).

##### Keep Final Figures In Millions (M), Rounded To 3 Decimals, To Maintain Consistency With Assets.

##### **Columns Considered**

| Group               | Columns                          |
| ------------------- | -------------------------------- |
| Master (To Drop)    | `Total_Liabilities`              |
| Housing             | `Housing_Loan`                   |
| Vehicle             | `Auto_Loan`                      |
| Personal / Consumer | `Personal_Loan`, `Consumer_Loan` |
| Revolving Credit    | `Credit_Card`                    |
| Education           | `Education_Loan`                 |
| Gold-Backed         | `Gold_Loan`                      |
| Others              | `Other_loan`                     |


##### Note:- Any Extra Aliases (E.g., “Consumer_Loan” = “Personal_loan”) Will Be Merged Into The Parent Bucket.

##### **Observation:-**

##### Total_Liabilities Is Just An Aggregation Of All Other Loan Columns → Safe To Drop.

##### Some Records Contain NaN, Blanks Or None; These Need To Be Filled With 0.

##### A Few Categories Overlap:

##### Personal_loan And Consumer_loan → Merged As Personal_loan.

##### No Duplicate For Housing, Auto, Credit Card, Education, Gold, Other.

##### **After Cleaning:-**

##### Every Loan Column Will Hold Its Value In M (E.g., 1250000 → 1.250 M).

##### A Single Row Total = Housing + Auto + Personal + Creditcard + Education + Gold + Other.

```python
# Cleaning & Merging Loans 

loan_cols = [
    "Total_Liabilities", "Housing_Loan", "Auto_Loan",
    "Personal_Loan", "Consumer_Loan", "Credit_Card",
    "Education_Loan", "Gold_Loan", "Other_loan"
]

# Fill Missing / Blank With 0

data[loan_cols] = data[loan_cols].replace(["", " ", "NaN", np.nan], 0).astype(float)

# Merge Consumer_Loan → Personal_Loan

data["Personal_Loan"] = data["Personal_Loan"] + data["Consumer_Loan"]

# Convert To Millions (3 Decimals)

for col in ["Housing_Loan", "Auto_Loan", "Personal_Loan",
            "Credit_Card", "Education_Loan", "Gold_Loan", "Other_loan"]:
    data[col] = data[col] / 1_000_000
    data[col] = data[col].round(3)

# Print Final DataFrame 

print("\n Cleaned Loan Data (in Millions):")
print(sorted(data[["Housing_Loan","Auto_Loan","Personal_Loan",
          "Credit_Card","Education_Loan","Gold_Loan","Other_loan"]].head()))
```

```python
# Drop Total_Liabilities & Consumer_Loan

data.drop(columns=["Total_Liabilities", "Consumer_Loan"], inplace=True)
```
